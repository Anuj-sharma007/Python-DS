```python
def binary_search(arr,key):
    left=0
    right=len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==key:
            return mid
        if key>arr[mid]:
            left=mid+1
        else:
            right=mid-1
    return -1
```

```python
binary_search([1,4,6,3,9],3)
```
