#### A Linked List is a fundamental linear data structure in computer science, similar to an array, but with a different memory structure. Instead of storing data sequentially in contiguous memory locations, it stores elements at arbitrary locations and uses pointers to connect them.

Structure and Components
A linked list is composed of a sequence of connected structures called nodes. Each node contains two main pieces of information:

Data: The actual value stored in the element (e.g., a number, a string, an object).

Pointer (or Link): A reference or address to the next node in the sequence.

```python
class Node():
    def __init__(self,data):
        self.data=data
        self.next=None
        
```

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4

current=node1
while current is not None:
    print(current.data,end="->")
    current=current.next
print("None")
```

## Adding nodes

#### adding new node at begining

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4

head=node1
new_node=Node(8)
new_node.next=head
head=new_node

current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("None")
```

#### adding new node at specific location

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4


new_node=Node(78)
head=node1
current=head
while current is not None and current.data!=5:
    current=current.next
new_node.next=current.next
current.next=new_node

head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("None")
```

#### adding new node at the end

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4
new_node=Node(10)

head=node1
current=head
while current.next is not None:
    current=current.next
current.next=new_node

head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")
```

## Deleting Nodes

#### deleting node from the begining

```python
head=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

head.next=node2
node2.next=node3
node3.next=node4


current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")

if head is not None:
    head=head.next

current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")


```

#### deleting node from the end

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4
head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")
head=node1
current=head
while current.next.next is not None:
    current=current.next
current.next=None

head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")
```

#### deleting a particular node

```python
node1=Node(3)
node2=Node(5)
node3=Node(7)
node4=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4
head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")
head=node1
current=head
while current.next.data!=7:
    current=current.next
current.next=current.next.next
head=node1
current=head
while current is not None:
    print(current.data,end="->")
    current=current.next
print("none")
```
