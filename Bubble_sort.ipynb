```python
%%time
def bubble_sort(arr):
    if isinstance (arr,(list,tuple)):
        n=len(arr)
        for i in range(n-1):
            for j in range(n-i-1):
                if arr[j]>arr[j+1]:
                    arr[j],arr[j+1]=arr[j+1],arr[j]
    else:
        print(f"Cannot perform on data type {type(arr)}")
                
    return arr 
```

```python
%%time
print(bubble_sort([79,78,45,84,99,25,86,68,230,56,75,97,14,84,24,78,86,34,66,22,77]))
```

```python
%%time
def bubble_sort_impro(arr):
    if isinstance (arr,(list,tuple)):
        n=len(arr)
        for i in range(n-1):
            swap=False
            for j in range(n-i-1):
                if arr[j]>arr[j+1]:
                    arr[j],arr[j+1]=arr[j+1],arr[j]
                    swap=True
            if swap==False:
                break
                
    return arr 
```

```python
%%time
print(bubble_sort_impro([79,78,45,84,99,25,86,68,230,56,75,97,14,84,24,78,86,34,66,22,77]))
```
