```python
def selection_sort(arr):
    if isinstance (arr,(list)):
        n=len(arr)
        for i in range(n-1):
            _min=i
            for j in range(i+1,n):
                if arr[j]<arr[_min]:
                    _min=j
            min_value=arr.pop(_min)
            arr.insert(i,min_value)
    return arr
```

```python
%%time
print(selection_sort([79,78,45,84,99,25,86,68,230,56,75,97,14,84,24,78,86,34,66,22,77]))
```

```python
def selection_sort_impro(arr):
    if isinstance (arr,(list)):
        n=len(arr)
        for i in range(n):
            _min=i
            for j in range(i+1,n):
                if arr[j]<arr[_min]:
                    _min=j
            arr[i],arr[_min]=arr[_min],arr[i]
                    
    return arr
```

```python
%%time
print(selection_sort_impro([79,78,45,84,99,25,86,68,230,56,75,97,14,84,24,78,86,34,66,22,77]))
```
