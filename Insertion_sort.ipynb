```python
%%time
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        insert_index=i
        current_value=arr.pop(i)
        for j in range(i-1,-1,-1):
            if arr[j]>current_value:
                insert_index=j
        arr.insert(insert_index,current_value)
    return arr
                
```

```python
%%time
insertion_sort([24,55,22,77,66,45,25,64,75,73,62])
```

```python
def insertion_sort_impro(arr):
    n=len(arr)
    for i in range(1,n):
        insert_index=i
        current_value=arr[i]
        for j in range(i-1,-1,-1):
            if arr[j]>current_value:
                arr[j+1]=arr[j]
                insert_index=j
            else:
                break
        arr[insert_index]=current_value
    return arr
                
```

```python
%%time
insertion_sort_impro([24,55,22,77,66,45,25,64,75,73,62])
```
